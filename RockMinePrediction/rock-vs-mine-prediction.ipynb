<h1 style="color:blue;text-align:center;">Importing the Dependencies</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

<h1 style="color:blue;text-align:center;">Data Collection and Data Preprocessing</h1>

```python
sonar_data = pd.read_csv("/kaggle/input/sonar-data/Sonar-Data.csv", header=None) #header = None is done because the header row is not there
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

## Getting the number of values of Rock and Mine

```python
sonar_data[60].value_counts() # to know number of Rock(R) and Mine(M), they are in 60th column
```

```python
sonar_data.groupby(60).mean()
```

## Separating Data and Label ('R' and 'M' column)

```python
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

<h1 style="color:blue;text-align:center;">Training and Test Data</h1>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=6)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

<h1 style="color:blue;text-align:center;">Model Training</h1>

```python
model = LogisticRegression()
```

## Training the Logistic Regression Model with training data

```python
model.fit(X_train, Y_train)
```

<h1 style="color:blue;text-align:center;">Model Evaluation</h1>

## Accuracy on training data

```python
XTrainPred = model.predict(X_train)
TrainDataAccuracy = accuracy_score(XTrainPred, Y_train)
TrainDataAccuracy
```

```python
XTestPred = model.predict(X_test)
TestDataAccuracy = accuracy_score(XTestPred, Y_test)
TestDataAccuracy
```
